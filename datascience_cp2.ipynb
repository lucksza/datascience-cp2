**Descrição da Base de Dados (RH)**

**A base contém informações detalhadas de colaboradores de uma empresa, incluindo dados demográficos, contratuais e comportamentais. São 36 colunas, entre elas:**


-Identificação e demografia: Nome, Gênero, Estado Civil, Raça, Idade.

-Contratuais: Departamento, Cargo, Salário Anual, Status de Emprego, Origem do Recrutamento.

-Indicadores de desempenho e satisfação: Nota de Desempenho, Engajamento (0–5), Satisfação (1–5), Faltas, Projetos Especiais, Atrasos.

-Histórico de RH: Data de contratação, data de desligamento (se houver), motivo da demissão, gestor responsável.

**Objetivo da análise**

**O objetivo é identificar padrões quantitativos que indiquem fatores críticos que afetam o engajamento e a satisfação dos funcionários. A análise buscará responder perguntas como:**

-Existe uma relação entre salário e satisfação?

-Funcionários com mais projetos especiais tendem a ter menos faltas ou maior engajamento?

-Há departamentos ou tipos de contratação mais propensos a baixa performance ou alta rotatividade?

-A partir desses insights, pretende-se propor melhorias em políticas de RH focadas em retenção de talentos, aumento de produtividade e redução de absenteísmo.
