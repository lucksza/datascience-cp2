# 3. Medidas descritivas


o Cálculo e análise de média, mediana, moda, variância, desvio
padrão e amplitude

o Gráficos (boxplots, histogramas, etc.) com interpretação

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('rh.csv')


variaveis = [
    'Salário (Anual)',
    'Engajamento (Escala de 0 a 5)',
    'Satisfação (Escala de 1 a 5)',
    'Projetos Especiais (Qtd.)',
    'Faltas (Total)',
    'Dias de Atraso nos Últimos 30 Dias',
    'ID da Nota de Desempenho'
]


plt.figure(figsize=(14, 10))
for i, col in enumerate(variaveis, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], kde=True, color="seagreen", bins=20)
    plt.title(f'{col}')
    plt.xlabel(col)
    plt.ylabel('Quantidade de Funcionários')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('rh.csv')

def calcularr(obj, coluna):
    media = round(df[coluna].mean(), 2)
    mediana = round(df[coluna].median(), 2)
    moda = int(df[coluna].mode()[0])
    variancia = round(df[coluna].var(), 2)
    desvio_padrao = round(df[coluna].std(), 2)
    amplitude = round(df[coluna].max() - df[coluna].min(), 2)

    print(f'{obj}')
    print(f'Média:         {media:,.2f}')
    print(f'Mediana:       {mediana:,.2f}')
    print(f'Moda:          {moda:,}')
    print(f'Variância:     {variancia:,.2f}')
    print(f'Desvio Padrão: {desvio_padrao:,.2f}')
    print(f'Amplitude:     {amplitude:,.2f}\n')

variaveis = [
    ('Salário (Anual)', 'Salário (Anual)'),
    ('Engajamento (0–5)', 'Engajamento (Escala de 0 a 5)'),
    ('Satisfação (1–5)', 'Satisfação (Escala de 1 a 5)'),
    ('Projetos Especiais', 'Projetos Especiais (Qtd.)'),
    ('Faltas (Total)', 'Faltas (Total)'),
    ('Dias de Atraso (30 dias)', 'Dias de Atraso nos Últimos 30 Dias'),
    ('Nota de Desempenho', 'ID da Nota de Desempenho')
]

for obj, coluna in variaveis:
    calcularr(obj, coluna)
```

# 4. Correlação


o Cálculo e visualização de correlação entre variáveis relevantes


o Interpretação contextualizada com o problema do grupo

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('rh.csv')

salary_by_sat = (
    df
    .groupby('Satisfação (Escala de 1 a 5)')['Salário (Anual)']
    .mean()
    .sort_index()
)

plt.figure(figsize=(6, 4))
plt.bar(
    salary_by_sat.index.astype(int),
    salary_by_sat.values,
    color='orange'
)
plt.title('Salário Médio por Nota de Satisfação')
plt.xlabel('Satisfação (1–5)')
plt.ylabel('Salário Médio ')
plt.xticks(salary_by_sat.index.astype(int))
plt.tight_layout()
plt.show()

salario_por_desempenho = (
    df
    .groupby('ID da Nota de Desempenho')['Salário (Anual)']
    .mean()
    .sort_index()
)


plt.figure(figsize=(8, 5))
plt.bar(
    salario_por_desempenho.index.astype(int),
    salario_por_desempenho.values,
    color='skyblue'
)
plt.title('Salário Médio por Nota de Desempenho')
plt.xlabel('Nota de Desempenho')
plt.ylabel('Salário Médio ')
plt.xticks(salario_por_desempenho.index.astype(int))
plt.tight_layout()
plt.show()


salario_por_projeto = (
    df.groupby('Projetos Especiais (Qtd.)')['Salário (Anual)']
    .mean()
    .sort_index()
)


plt.figure(figsize=(8, 5))
plt.bar(
    salario_por_projeto.index,
    salario_por_projeto.values,
    color='mediumseagreen',
    edgecolor='black'
)
plt.title('Salário Médio por Quantidade de Projetos Especiais')
plt.xlabel('Projetos Especiais (Qtd.)')
plt.ylabel('Salário Médio ')
plt.xticks(salario_por_projeto.index)
plt.tight_layout()
plt.show()



import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('rh.csv')


df_renomeado = df.rename(columns={
    'satisfacao': 'Satisfação (Escala de 1 a 5)',
    'salario_anual': 'Salário (Anual)',
    'nota_desempenho': 'ID da Nota de Desempenho',
    'projetos_especiais': 'Projetos Especiais (Qtd.)'
})


df_corr = df_renomeado[[
    'Satisfação (Escala de 1 a 5)',
    'Salário (Anual)',
    'ID da Nota de Desempenho',
    'Projetos Especiais (Qtd.)'
]]


correlacoes = df_corr.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(
    correlacoes,
    annot=True,
    cmap="Greens",
    fmt=".2f",
    linewidths=0.5,
    cbar=True
)
plt.title("Correlação: Satisfação, Salário, Desempenho e Projetos Especiais", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

**1. Relação entre Salário e Satisfação**


O gráfico “Salário Médio por Nota de Satisfação” mostra uma tendência positiva entre a satisfação e o salário médio. Funcionários com notas de satisfação mais altas (principalmente notas 4 e 5) apresentam salários médios mais elevados do que aqueles com notas mais baixas (1 e 2).

O que isso sugere?

Existe sim uma correlação direta entre salário e satisfação. Funcionários mais satisfeitos tendem a receber mais.

Essa relação pode refletir dois aspectos:

Colaboradores mais bem remunerados se sentem mais valorizados, o que aumenta a satisfação.

Ou, inversamente, funcionários mais engajados e produtivos (logo, mais satisfeitos) são recompensados com salários maiores.

---

**2. Relação entre Salário e Nota de Desempenho**


No gráfico “Salário Médio por Nota de Desempenho”, observamos uma escala ascendente clara: quanto maior a nota de desempenho, maior é o salário médio. Funcionários com nota 4 (mais alta no conjunto analisado) apresentam o maior salário médio de todos os grupos.

O que isso sugere?

A empresa remunera com base no desempenho individual, o que é uma boa prática de meritocracia.

Essa estrutura pode também funcionar como um incentivo indireto à performance — funcionários sabem que, ao entregarem mais resultados, podem ser recompensados.


---

**3. Relação entre Salário e Projetos Especiais(Alta visibilidade interna,Curto a médio prazo)**


Funcionários com 0 ou 1 projeto especial possuem salários médios abaixo de R$65.000.

A partir de 2 projetos, há uma elevação clara, chegando perto ou ultrapassando os R$90.000.

Os salários tendem a se estabilizar ou crescer moderadamente a partir de 3 projetos, indicando uma possível recompensa proporcional ao nível de envolvimento.
